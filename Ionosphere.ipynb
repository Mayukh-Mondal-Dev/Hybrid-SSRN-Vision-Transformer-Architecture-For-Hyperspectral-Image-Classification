```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt

# Load MATLAB file
mat_data = scipy.io.loadmat('ionosphere.mat')

# Assuming the variables are stored as 'act', 'feat', and 'label'
act = mat_data['act']
feat = mat_data['feat']
label = mat_data['label']

# If 'act' is categorical, create a bar plot to visualize its distribution
unique_act, act_counts = np.unique(act, return_counts=True)
indices = np.arange(len(unique_act))  # Generate numerical indices for categories
plt.figure()
plt.bar(indices, act_counts)
plt.xticks(indices, unique_act)  # Set category names as x-axis ticks
plt.title('act Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Plotting 'feat'
plt.figure()
plt.plot(feat)
plt.title('feat')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

# If 'label' is categorical, you might want to count occurrences of each class
unique_labels, label_counts = np.unique(label, return_counts=True)
plt.figure()
plt.bar(unique_labels, label_counts)
plt.title('Label Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
