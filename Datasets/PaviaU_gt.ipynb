```python
import scipy.io
import matplotlib.pyplot as plt

# Load MATLAB file
mat_data = scipy.io.loadmat('PaviaU_gt.mat')

# Assuming the ground truth map is stored in a variable named 'indian_pines_gt'
ground_truth_map = mat_data['paviaU_gt']
print("Ground truth map shape:", ground_truth_map.shape)

# Visualize the ground truth map
plt.figure()
plt.imshow(ground_truth_map, cmap='jet')  # Assuming categorical labels, using 'jet' colormap
plt.title('Ground Truth Map')
plt.colorbar(label='Class')
plt.axis('off')
plt.show()
```
