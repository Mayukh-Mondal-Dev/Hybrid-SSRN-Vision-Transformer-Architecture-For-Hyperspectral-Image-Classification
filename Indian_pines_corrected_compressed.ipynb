```python
import scipy.io
import matplotlib.pyplot as plt

# Load MATLAB file
mat_data = scipy.io.loadmat('Indian_pines_corrected_compressed.mat')

# Assuming the images are stored in a variable named 'images'
images = mat_data['indian_pines']

# Visualize each image
for i in range(images.shape[2]):
    plt.figure()
    plt.imshow(images[:, :, i], cmap='gray')  # Assuming grayscale images
    plt.title('Image {}'.format(i+1))
    plt.axis('off')
    plt.show()
```
